# Types of Data
### 1) Numeric:
Numerical data represents values that are measurable and can be quantified.
It can be used for mathematical operations such as addition, subtraction, multiplication, or division.
It is typically represented using graphs like histograms, line graphs, or scatter plots.
### Types of Numerical Data:
#### Continuous Data:
Can take any value within a range.
Examples: Height (175.5 cm), Temperature (36.7°C), Weight (73.5 kg).
#### Discrete Data:
Consists of specific, countable values.
Examples: Number of students
### 1) Non Numeric Data
Non-numerical data, also known as categorical data, represents labels or characteristics 
It cannot be used for direct mathematical calculations.
It is often analyzed using bar charts, pie charts, or frequency tables.
### Label: 
The label is the output or target variable that the machine learning model is trying to predict or classify.
A label is the outcome or target variable in a dataset that you are trying to predict or analyze.
* Example:
In a dataset predicting house prices:

Input features: Size of the house, number of rooms, location.
Label: Actual price of the house.

Labels make charts and graphs easier to interpret by adding context to the data.
### Feature: 
 Features are used as input variables in data analysis, machine learning, and statistics to build predictive or descriptive models.
 * eg:
#### FOR CLASSIFICATION TASK
 * Image Classification
Task: Classify images as Cat or Dog.
Features:
Color histogram (distribution of colors).
Shape descriptors (edges or contours).
Texture patterns (smooth vs. rough areas).
#### FOR REGRESSION TASK
Weather Forecasting
Task: Predict tomorrow’s temperature.
Features:
Today’s temperature.
Humidity levels.
Wind speed.
Cloud cover percentage.
## Exploratory data analysis:
Exploratory Data Analysis (EDA) is a critical step in the data analysis process. It involves analyzing datasets to summarize their main characteristics, often with the help of visualizations and statistical techniques. EDA helps uncover patterns, detect anomalies, test hypotheses, and check assumptions about the data.
### Why is EDA Important?
Identifies Trends: Helps understand distributions, patterns, and relationships.
Detects Errors: Spots missing values, outliers, and anomalies in data.
Guides Modeling: Identifies which features are important and what preprocessing is required.
Builds Intuition: Helps analysts understand the data better before applying machine learning models.
### Objectives of EDA:
* Understand the Data Structure:
Get a sense of the data, including its size, types of variables, and overall structure.
* Summarize Main Characteristics:
Calculate descriptive statistics like mean, median, mode, variance, and standard deviation.
* Detect Missing Values and Outliers:
Identify any missing or unusual values that may require cleaning or imputation.
* Examine Relationships:
Understand the relationships between features (independent variables) and the label (dependent variable) if applicable.
* Assess Data Distribution:
Check the distribution of variables to decide on transformations, scaling, or normalization if required.
## Steps in EDA:

1. * Understand the Dataset: Load the data, inspect its structure, and check data types and dimensions.  
2. * Check for Missing Values and Duplicates: Identify and handle missing or redundant data by imputation or removal.  
3. * Perform Univariate Analysis: Analyze individual variables using summary statistics and visualizations.  
4. * Conduct Bivariate Analysis: Explore relationships between two variables using scatter plots, boxplots, and correlation analysis.  
5. * Perform Multivariate Analysis: Examine interactions among multiple variables using pair plots, heatmaps, or dimensionality reduction.  
6. * Visualize Data: Use graphs and plots to uncover patterns, trends, and anomalies.  
7. * Prepare for Modeling: Clean and transform the data based on insights for further analysis or modeling.
### Tools for EDA:
Python Libraries: Pandas, NumPy, Matplotlib, Seaborn, Plotly.
R Libraries: dplyr, ggplot2, tidyr.
Specialized Tools: Tableau, Power BI.
## Visualizations for Different Variables:

# feature extraction and transformation

```python
from scipy import stats
import matplotlib.pyplot as plt
data = np.random.exponential(scale=1, size = 1000)
transformed_data, lamda_value = stats.boxcox(data)

plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.hist(data, bins=30)
plt.title('Orginal Data')

plt.subplot(1,2,2)
plt.hist(transformed_data, bins=30)
plt.title("Box-cox transformed data")
plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
data= np.random.exponential(scale=1,size=1000)
transfromed_data, lambda_value= stats.boxcox(data)
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.hist(data, bins=30,color='y')
plt.title('original Data',color='Blue' )
plt.subplot(1,2,2)
plt.hist(transfromed_data, bins=30, color='g')
plt.title('Box-cox Transformed Data',color='Blue')
plt.show()
```

## Handling text data
### Bag of words:  
*  Representing text data in numerical form so that it can be used for machine learning and natural language processing (NLP) tasks.
* It simplifies the text data by treating it as a collection (or "bag") of individual words, ignoring grammar and word order while preserving the frequency or presence of words.
### Text Data
1.   Bag of Words (BoW)
2.   Term Frequency-Inverse Document Frequency (TF-IDF)
3.   Word2Vec (W2V)
4.   Average Word2Vec (AvgW2V)
5.   TF-IDF weighted Word2Vec
